```python
a=df.Payment.value_counts()
print(a)
print("total: ",df.Payment.count())
```

```python
quarterly_totals = df.groupby('Quarter')['Total'].sum()
print(quarterly_totals)
```

```python
plt.plot(quarterly_totals.index, quarterly_totals.values, marker='o')

plt.xlabel('Quarter')
plt.ylabel('Total')
plt.title('Quarterly Totals')

plt.show()
```

```python
Branch_totals = df.groupby('Branch')['Total'].sum()
print(Branch_totals)
```

```python
plt.bar(Branch_totals.index, Branch_totals.values)

plt.xlabel('Branch')
plt.ylabel('Total')
plt.title('Branch Totals')

plt.show()
```

```python
Products_totals = df.groupby('Product')['Quantity'].sum()
print(Products_totals)
```

```python
plt.bar(Products_totals.index, Products_totals.values)

plt.xlabel('Product')
plt.ylabel('Total Quantity')
plt.title('Product Quantity Totals')

plt.xticks(rotation=80)

plt.show()
```

```python
branch_a_df = df[df['Branch'] == 'A']
branch_b_df = df[df['Branch'] == 'B']
branch_c_df = df[df['Branch'] == 'C']

customer_counts_A = branch_a_df['Customer'].value_counts()
customer_counts_B = branch_b_df['Customer'].value_counts()
customer_counts_C = branch_c_df['Customer'].value_counts()

print("Branch_A")
print(customer_counts_A)
print("Branch_B")
print(customer_counts_B)
print("Branch_C")
print(customer_counts_C)
```

```python
branches = ['Branch A', 'Branch B', 'Branch C']
member_counts = [167, 165, 169]
normal_counts = [173, 167, 159]

bar_width = 0.35

r1 = range(len(branches))
r2 = [x + bar_width for x in r1]

plt.bar(r1, member_counts, width=bar_width, label='Member')
plt.bar(r2, normal_counts, width=bar_width, label='Normal')

plt.xlabel('Branch')
plt.ylabel('Count')
plt.title('Customer Counts by Branch and Type')
plt.xticks([r + bar_width/2 for r in range(len(branches))], branches)
plt.legend()

plt.show()
```

```python
avg_rating_by_branch = df.groupby('Branch')['Rating'].mean()
print(avg_rating_by_branch)
```

```python
branch_a_df = df[df['Branch'] == 'A']
branch_b_df = df[df['Branch'] == 'B']
branch_c_df = df[df['Branch'] == 'C']

Product_counts_A = branch_a_df['Product'].value_counts()
Product_counts_B = branch_b_df['Product'].value_counts()
Product_counts_C = branch_c_df['Product'].value_counts()

print("Branch_A")
print(Product_counts_A)
print("Branch_B")
print(Product_counts_B)
print("Branch_C")
print(Product_counts_C)
```

```python
products = ['Home and lifestyle', 'Electronic accessories', 'Sports and travel', 'Food and beverages', 'Fashion accessories', 'Health and beauty']
branch_a_counts = [65, 60, 59, 58, 51, 47]
branch_b_counts = [50, 55, 62, 50, 62, 53]
branch_c_counts = [45, 55, 45, 66, 65, 52]

bar_width = 0.2

r1 = range(len(products))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, branch_a_counts, width=bar_width, label='Branch A')
plt.bar(r2, branch_b_counts, width=bar_width, label='Branch B')
plt.bar(r3, branch_c_counts, width=bar_width, label='Branch C')

plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Product Counts by Branch')
plt.xticks([r + bar_width for r in range(len(products))], products, rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
